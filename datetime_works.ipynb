```python
start='2019-12-31 23:59:00'
end='2020-01-01 00:00:30'
def timedata(start, end):
    # Kirjoita vastauksesi tähän alle:
    import pandas as pd
    import numpy as np
    df=pd.DataFrame(pd.date_range(start ,end, freq="20S"))
    df['t,s']=df[0]-df[0][0]
    df['t,s']=df['t,s'].dt.total_seconds()
    df["hh.mm.ss"]=pd.to_datetime(df[0] ,unit='s')
    df["hh.mm.ss"]= df["hh.mm.ss"].dt.strftime("%H.%M.%S")
    df.set_index("hh.mm.ss", inplace=True)
    t = np.array(df['t,s'])
    x=(t*600)/(2**0.5)
    df["x,m"]=x
    y=(t*600)/(2**0.5)-(9.82*np.power(t,2))/2
    df["y,m"]=y
    df=df[['t,s',"x,m","y,m"]]
    tulos=df.applymap(lambda x: '{:.0f}'.format(x))
    # Älä kirjoita tähän muita print komentoja ennen viimeistä print(tulos) komentoa, koska tällöin vastauksesi saatetaan tulkita väärin.
    
    # Lopuksi vastauksesi tulostetaan tulos muuttujassa
    print(tulos) 
    
timedata(start, end)
```

```python
url='https://student.labranet.jamk.fi/~huoptu/datananal_s20/kerta6/harjoitukset/motion.csv'

def interp_times(url):
    # Kirjoita vastauksesi tähän alle:
    import pandas as pd
    df=pd.read_csv(url, sep='\t',decimal=".")
    df["hh:mm:ss"]=pd.to_datetime(df["hh:mm:ss"] )
    df['t_tot, s']=df["hh:mm:ss"]-df["hh:mm:ss"].iloc[0]
    df['t_tot, s']=df['t_tot, s'].dt.total_seconds()
    df.set_index("hh:mm:ss", inplace=True)
    df=df.resample('5S')[["y, m","t_tot, s"]].mean()
    df['t_tot, s']=df['t_tot, s'].interpolate()
    df["int_y, m"]=df["y, m"].interpolate()
    df1=df[["t_tot, s","int_y, m"]]
    df1=df1[("00:00:45"<= df1.index )&(df1.index<='00:01:31')]
    df1.index = pd.to_datetime(df1.index, unit="s").strftime("%H:%M:%S")
    df1.index.name = "hh:mm:ss"
    tulos=df1.applymap(lambda x: '{:.0f}'.format(x))
    
    # Älä kirjoita tähän muita print komentoja ennen viimeistä print(tulos) komentoa, koska tällöin vastauksesi saatetaan tulkita väärin.
    
    # Lopuksi vastauksesi tulostetaan tulos muuttujassa
    print(tulos)
    
interp_times(url)
```

## 5. Aikasarjojen yhdistäminen

Tässä tehtävässä yhdistetään kaksi eri aikasarjadatasettiä niiden aikaleimojen perusteella.

Datasetissä nro 1. aikainformaatio on sisällytetty ensimmäiseen ensimmäiseen nimeämättömään sarakkeeseen muodossa 'vuosi-kuukausi-päivä tunnit:minuutit:sekunnit'. Datasetissä nro 1. tarvitaan jatkolaskuja varten sarakkeen 'Air temperature, degC' lämpötilainformaatio.

Datasetissä nro 2. aikainformaatio on sisällytetty ensimmäiseen sarakkeeseen 'time, s' sekunteina. Datasetissä nro 2. tarvitaan jatkolaskuja varten sarakkeen 'Temp, C-deg' lämpötilainformaatio.

Tee funktio tempdiff(url1, url2), joka tekee seuraavaa:

lataa aikasarjadatan nro 1. DataFrameen url:sta ja aikasarjadatan nro 2. toiseen DataFrameen url:sta,

muodostaa datan nro 2. DataFrameen aikaleiman indeksiksi muodossa 'vuosi-kuukausi-päivä tunnit:minuutit:sekunnit', kun mittaukset ovat alkaneet '2020-11-05 08:40:00' ja sarake 'time, s' kertoo ajan mittausten alusta sekunteina,

yhdistää aikasarjadatojen nro 1. ja nro 2. kaikki lämpötilat 'Air temperature, degC' ja 'Temp, C-deg' yhteen 

DataFrameen aikaleimojen perusteella kasvavaan aikajärjestykseen siten, että yhdistetyssä DataFramessa indeksi kertoo mittaustapahtumien ajanhetken,

puuttuvien lämpötilojen arvot täytetään DataFrameen em. aikajärjestyksessä edellisillä lämpötilan arvoilla,

laskee uuteen DataFrameen sarakkeeseen 'Temp_diff, C-deg' kuinka paljon lämpötila 'Temp, C-deg' on suurempi kuin lämpötila 'Air temperature, degC',

laskee DataFrameen lämpötilojen erotuksen 'Temp_diff, C-deg' liukuvan keskiarvon kolmen tunnin aikaikkunalla sarakkeeseen 'Temp_roll, C-deg',

pyöristää ja esittää DataFramen lukuarvot kolmen desimaalin tarkkuudella,

tulostaa lopputulokseen viisi ensimmäistä riviä sarakkeista 'Temp_diff, C-deg' ja 'Temp_roll, C-deg' ajanhetkestä '2020-11-05 12:59:00' eteenpäin.

```python
def tempdiff(url1, url2):
    # Kirjoita vastauksesi tähän alle:
    import pandas as pd
    df1=pd.read_csv(url1)
    df2=pd.read_csv(url2, sep="\t", decimal=".")
    x=pd.to_datetime('2020-11-05 08:40:00')
    time= x+pd.to_timedelta(df2['time, s'],'s')
    df3=df2[["Temp, C-deg"]]
    df3=df3.reset_index()
    df3=df3.set_index(time)
    df1=df1.set_index("Unnamed: 0")
    df1=df1["Air temperature, degC"]
    
    df = df3.merge(df1, left_index=True, right_index=True, how="outer")
    
    df=df.drop(columns="index")
    df=df.ffill()
    df["Temp_diff, C-deg"]=df["Temp, C-deg"]-df["Air temperature, degC"]
    df["Temp_roll, C-deg"]=df["Temp_diff, C-deg"].rolling("3h").mean()
    df=df[["Temp_diff, C-deg","Temp_roll, C-deg"]].applymap(lambda x: '{:.3f}'.format(x))
    df=df[df.index > '2020-11-05 12:59:00' ]
    tulos= df.rename_axis(index=None).head()
    # Älä kirjoita tähän muita print komentoja ennen viimeistä print(tulos) komentoa, koska tällöin vastauksesi saatetaan tulkita väärin.
    
    # Lopuksi vastauksesi tulostetaan tulos muuttujassa
    print(tulos)

# Voit kokeillä funktiosi toimivuutta:
url1 = 'https://student.labranet.jamk.fi/~huoptu/datananal_s20/kerta6/harjoitukset/saatiedot_kumpula_1d.csv'    
url2 = 'https://student.labranet.jamk.fi/~huoptu/datananal_s20/kerta6/harjoitukset/indoormeas.csv'

tempdiff(url1, url2)
```
