## harjoitus 1 Päästöjen aiheuttajia

Tee funktio emissions(url), joka tekee järjestyksessä seuraavaa:

lataa datan DataFrameen url:sta, jossa erotinmerkkinä on tabulaattori ('\t'),

tallentaa DataFrameen sarakkeet 'country' ja 'main activity sector name',

poistaa DataFramesta rivit, jos sarakkeessa 'main activity sector name' esiintyy merkkijonot '20-99 All stationary installations' tai '21-99 All industrial installations (excl. combustion)'

```python
def emissions(url):
    import pandas as pd

    df = pd.read_csv(url, sep='\t')
    sarakkeet = 'country\tcountry_code\tETS information\tmain activity sector name\tunit\tvalue\tyear'
    country = df[sarakkeet].str.split('\t').str[0]
    sector = df[sarakkeet].str.split('\t').str[3]

    df2 = pd.DataFrame({'country': country, 'sector': sector})
    
    str1 = '20-99 All stationary installations'
    str2 = '21-99 All industrial installations (excl. combustion)'
    df2=df2[~df2['sector'].isin([str1, str2])]
    
    tulos = df2['sector'].value_counts(ascending=True).head(5)

    print(tulos)
    
emissions("https://student.labranet.jamk.fi/~huoptu/datananal_s20/kerta3/harjoitukset/ETS_Database_v38A.csv")
```

```python
def emissions(url):
    import pandas as pd

    df = pd.read_csv(url, sep='\t')
    sarakkeet = 'country\tcountry_code\tETS information\tmain activity sector name\tunit\tvalue\tyear'
    sector = df[sarakkeet].str.split('\t').str[3]
    value = df[sarakkeet].str.split('\t').str[5]
    year = df[sarakkeet].str.split('\t').str[6]
    
    df2 = pd.DataFrame({'sector': sector, 'value': value, 'year': year})
    
    str1 = '20-99 All stationary installations'
    str2 = '21-99 All industrial installations (excl. combustion)'
    df2=df2[~df2['sector'].isin([str1, str2])]
    
    df2['value']=pd.to_numeric(df2['value'], errors='coerce')
    df2['year']=pd.to_numeric(df2['year'], errors='coerce')
    df2=df2[(df2['year']<=2015) &(df2['year']>=2010)]
    
    tulos = df2.groupby(df2['sector'], as_index=False)['value'].sum().sort_values('value', ascending=False).head(5)
    tulos = tulos[['value', 'sector']]
    
    print(tulos)
    
emissions("https://student.labranet.jamk.fi/~huoptu/datananal_s20/kerta3/harjoitukset/ETS_Database_v38A.csv")
```

```python
url="https://student.labranet.jamk.fi/~huoptu/datananal_s20/kerta3/harjoitukset/ETS_Database_v38A.csv"

def emissions(url):
    import pandas as pd

    df = pd.read_csv(url, sep='\t')
    sarakkeet = 'country\tcountry_code\tETS information\tmain activity sector name\tunit\tvalue\tyear'
    sector = df[sarakkeet].str.split('\t').str[3]
    value = df[sarakkeet].str.split('\t').str[5]
    year = df[sarakkeet].str.split('\t').str[6]
    ccode = df[sarakkeet].str.split('\t').str[1]
    
    df2 = pd.DataFrame({'ccode': ccode, 'sector': sector, 'value': value, 'year': year})
    
    df2['value']=pd.to_numeric(df2['value'], errors='coerce')
    df2['year']=pd.to_numeric(df2['year'], errors='coerce')
    df2=df2[(df2['year']<=2015) &(df2['year']>=2010)]
    df2['year']=df2['year'].astype(int)
    
    str1 = '20-99 All stationary installations'
    str2 = '21-99 All industrial installations (excl. combustion)'
    df2=df2[~df2['sector'].isin([str1, str2])]
    
    df2 = df2.sort_values(['year'], ascending=True)
    summat = df2['value'].groupby(df2['year']).sum()
    
    tulos = pd.DataFrame({'emissions': summat})
    
    tulos['diff'] = tulos['emissions'].diff()
    tulos['change']=((tulos['diff'])/(tulos['emissions'].shift(1))*100)
    
    tulos['cumsum']=tulos['emissions'].cumsum()
    tulos=tulos.drop('diff', axis=1)
    
    print(tulos)
    
emissions("https://student.labranet.jamk.fi/~huoptu/datananal_s20/kerta3/harjoitukset/ETS_Database_v38A.csv")
```

```python
def emissions(url):
    import pandas as pd

    df = pd.read_csv(url, sep='\t')
    sarakkeet = 'country\tcountry_code\tETS information\tmain activity sector name\tunit\tvalue\tyear'
    sector = df[sarakkeet].str.split('\t').str[3]
    value = df[sarakkeet].str.split('\t').str[5]
    year = df[sarakkeet].str.split('\t').str[6]
    ccode = df[sarakkeet].str.split('\t').str[1]
    
    df2 = pd.DataFrame({'ccode': ccode, 'sector': sector, 'value': value, 'year': year})
    
    df2['value']=pd.to_numeric(df2['value'], errors='coerce')
    df2['year']=pd.to_numeric(df2['year'], errors='coerce')
    df2=df2[(df2['year']<=2015) &(df2['year']>=2010)]
    df2['year']=df2['year'].astype(int)
    
    str1 = '20-99 All stationary installations'
    str2 = '21-99 All industrial installations (excl. combustion)'
    df2=df2[~df2['sector'].isin([str1, str2])]
    
    df3 = df2.groupby([df2['ccode'],df2['year']], as_index=False)['value'].sum() # tuloksena tulee suoraan DataFrame!
    
    df4 = df3.groupby('ccode', as_index=False)['value'].max()
    booleans = (df3['ccode'].isin(df4['ccode']) & df3['value'].isin(df4['value']))
    tulos = df3.loc[booleans,:].sort_values('value', ascending=False).head(5)
    print(tulos)
    

emissions("https://student.labranet.jamk.fi/~huoptu/datananal_s20/kerta3/harjoitukset/ETS_Database_v38A.csv")
```

```python
def emissions(url):
    import pandas as pd
    import numpy as np
    pd.options.mode.chained_assignment = None
    def poistaalusta(x):
            return x[1:]

    def poistalopusta(x):
            return x[:-1]

    df = pd.read_table(url,sep='\t',quotechar = '.', converters = {'"country' : poistaalusta, 'year"' : poistalopusta})
    df["year"]=df['year"']
    df=df[["country_code","main activity sector name","value","year"]]
    l=(df["main activity sector name"]!='20-99 All stationary installations') & (df["main activity sector name"]!= "21-99 All industrial installations (excl. combustion)")
    df=df.loc[l]
    df2=df[(df["year"].str.isnumeric())]
    df2["year"]=df2["year"].astype(int)
    lista=(df2["year"]>= 2010) & (df2["year"]<= 2015)
    df2=df2[lista]
    df3 = pd.DataFrame(df2['value'].groupby([df2['country_code'], df2['year']]).sum())
    df3=df3.reset_index()
    df4=df3.loc[(df3.groupby('country_code')['value'].idxmax())]
    index=np.array(df4.index)
    df5=df3.drop(index,axis=0)
    df5=df5.loc[(df5.groupby('country_code')['value'].idxmax())]
    df5=df5.sort_values(by="value",ascending=False)
    df5["ccode"]=df5["country_code"]
    tulos=df5[["ccode","year","value"]].head()
    print(tulos)
    
# Testaa funktiotasi:
url='https://student.labranet.jamk.fi/~huoptu/datananal_s20/kerta3/harjoitukset/ETS_Database_v38A.csv'
emissions(url)
```

```python
def emissions(url):
    import pandas as pd

    df = pd.read_csv(url, sep='\t')
    sarakkeet = 'country\tcountry_code\tETS information\tmain activity sector name\tunit\tvalue\tyear'
    sector = df[sarakkeet].str.split('\t').str[3]
    value = df[sarakkeet].str.split('\t').str[5]
    year = df[sarakkeet].str.split('\t').str[6]
    ccode = df[sarakkeet].str.split('\t').str[1]
    
    df2 = pd.DataFrame({'ccode': ccode, 'sector': sector, 'value': value, 'year': year})
    
    df2['value']=pd.to_numeric(df2['value'], errors='coerce')
    df2['year']=pd.to_numeric(df2['year'], errors='coerce')
    df2=df2[(df2['year']<=2015) &(df2['year']>=2010)]
    df2['year']=df2['year'].astype(int)
    
    str1 = '20-99 All stationary installations'
    str2 = '21-99 All industrial installations (excl. combustion)'
    df2=df2[~df2['sector'].isin([str1, str2])]
    
    df3 = df2.groupby([df2['ccode'],df2['year']], as_index=False)['value'].sum() # tuloksena tulee suoraan DataFrame!
    df3 = df3.sort_values(['ccode', 'value'], ascending=False)
    
    def toinen(x):
        return x.iloc[1]
    
    df4 = df3.groupby('ccode', as_index=False)['value'].agg(toinen)
    booleans = (df3['ccode'].isin(df4['ccode']) & df3['value'].isin(df4['value']))
    tulos = df3.loc[booleans,:].sort_values('value', ascending=False).head(5)
    
    print(tulos)
```
